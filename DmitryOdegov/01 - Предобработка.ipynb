```python
import shutil
import pandas as pd
from datetime import date
```

```python
file_in = 'dataset1_02.xlsx'
file_out_xls = 'output.xlsx'
file_out_csv = 'output.csv'
shutil.copy(f'/content/drive/MyDrive/UII/Fintex/{file_in}', f'/content/{file_in}')
```

```python
excel_file = pd.ExcelFile(file_in)
```

#Предобработка датасета

```python
# Параметры предобработки

min_amount = 1000       # минимальная сумма кредита
min_year_credit = 2004  # минимальный год выданного кредита

print('Количество строк до обработки:', df.shape[0])

# удаление строк по условиям
df = df.drop(df[df.issued.dt.year < min_year_credit].index)  # минимальная дата выданного кредита
df = df.drop(df[df.amount < min_amount].index)               # минимальная сумма кредита

# удаление строк с пустым значением в определённых столбцах
df = df.drop(df[df.type_credit==''].index)    # тип кредита
df = df.drop(df[df.status_credit==''].index)  # статус кредита

# удаление строк с недопустимой валютой (не RUR)
df = df.drop(df[df.currency!='RUB'].index)

print('Количество строк после обработки:', df.shape[0])
```

#Сохранение обработанного датасета в XLS и CSV

```python
df.to_excel(file_out_xls)
shutil.copy(f'/content/{file_out_xls}', f'/content/drive/MyDrive/UII/Fintex/{file_out_xls}')
```

```python
df.to_csv (file_out_csv, index=False)
shutil.copy(f'/content/{file_out_csv}', f'/content/drive/MyDrive/UII/Fintex/{file_out_csv}')
```
