# Сервисные функции

```python
# Получить инфо
def get_info(data):
    print('Размеры датасета', data.shape)
    print('Общее количество пропусков:', data.isna().sum().sum(), 'или', round(data.isna().sum().sum() / (data.shape[0]*data.shape[1])*100, 2), '%\n')
    print(f'Доли пропусков по столбцам:\n{round(data.isna().sum() / data.shape[0] * 100, 4)}')
    print(data.info())
    round(data.describe(), 1)

# Добавить столбец target
def add_target_column(data, cls: int):
    target_col = [cls for i in range(data.shape[0])]
    data.insert(loc=len(data.columns) , column='target', value=target_col)
    return data

# Обьединить таблицы
def merge_tables(tables):
    if len(tables) == 0:
        return None
    result = pd.concat(tables)
    return result
```

# Предобработка данных

```python
# Загружаем данные
data_1 = pd.read_excel(f'{path_to_folder}/2567_КИдлятеста_5000_безПИ_1_класс.xlsx', header=(2))
# data_2 = pd.read_excel(f'{path_to_folder}/', header=(2))
data_3 = pd.read_excel(f'{path_to_folder}/2581_КИдлятеста_всего1727_безПИ_3_класс.xlsx', header=(2))
data_4 = pd.read_excel(f'{path_to_folder}/2579_КИдлятеста_плохие_всего4588_безПИ_4_класс.xlsx', header=(2))

# print(data_1.shape[0] + data_2.shape[0] + data_3.shape[0] + data_4.shape[0])
print(data_1.shape[0] + data_3.shape[0] + data_4.shape[0])

# Добавляем столбец target
data_1 = add_target_column(data_1, 1)
# data_2 = add_target_column(data_2, 2)
data_3 = add_target_column(data_3, 3)
data_4 = add_target_column(data_4, 4)

# Объединеняем таблицы
# data = merge_tables([data_1, data_2, data_3, data_4])
data = merge_tables([data_1, data_3, data_4])

# Заменяем столбцы на русские
data.columns = rus_cols

get_info(data)
data.head()
```

```python
df = data.copy()
```

```python
# Параметры предобработки
min_year_credit = 2004                                                          # Минимальный год выданного кредита
min_amount = 1000                                                               # Минимальная сумма кредита

print('Количество строк до обработки:', df.shape[0])

# Удаление строк по условиям
df = df.drop(df[df['Выдан'].dt.year < min_year_credit].index)                   # Минимальная дата выданного кредита
df = df.drop(df[df['Сумма'] < min_amount].index)                                # Минимальная сумма кредита

# Удаление строк с пустым значением в определённых столбцах
df = df.drop(df[df['Тип_кредита']==''].index)                                   # Тип кредита
df = df.drop(df[df['Статус']==''].index)                                        # Статус кредита

# Удаление строк с недопустимой валютой (не RUR)
df = df.drop(df[df['Валюта']!='RUB'].index)
df = df.drop(['Валюта'], axis=1)

# # Преобразование столбца "Срок" в тип данных datetime64[ns]
# df['Срок'] = pd.to_datetime(df['Срок'], errors='coerce')

# Удаление строк с недопустимыми датами (NaT)
df = df.dropna(subset=['Срок'])

print('Количество строк после обработки:', df.shape[0])
```

1. Надо ввести соотношение суммы просрочек по отношению к обьему кредита
2. Как вариант кол-во дней просрочки к кол-ву дней кредита
3. Так же если есть данные о погашении после напоминания или самостоятельно гасил

```python
# 1. ПЗ - просроченная задолженность

df['ПЗ_по_займу/Сумма'] = df['Просроченная_задолженность_по_займу'] / df["Сумма"]
df['МаксПЗ_по_займу/Сумма'] = df['Максимальная_сумма_просроченной_задолженности'] / df["Сумма"]
```
